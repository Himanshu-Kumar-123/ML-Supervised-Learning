```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pickle
# Load data (deserialize)

with open('preprocessed_X_sm.pickle', 'rb') as handle:
    X_train = pickle.load(handle)

with open('X_test.pickle', 'rb') as handle:
    X_test = pickle.load(handle)

with open('y_sm.pickle', 'rb') as handle:
    y_train = pickle.load(handle)

with open('y_test.pickle', 'rb') as handle:
    y_test = pickle.load(handle)
```

```python
X_train.head()
```

```python
y_train.head()
```

## Sklearn Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(random_state=7, max_depth=1)
tree_clf.fit(X_train, y_train)

print("Train accuracy: {:.2f}".format(tree_clf.score(X_train, y_train)*100))
print("Test accuracy: {:.2f}".format(tree_clf.score(X_test, y_test)*100))
```

## Sklearn Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=7, max_depth= 6)
model.fit(X_train, y_train)
```

```python
print("Train accuracy: {:.2f}".format(model.score(X_train, y_train)*100))
print("Test accuracy: {:.2f}".format(model.score(X_test, y_test)*100))
```
